```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ours = pd.read_csv('rebuttal/seu.csv')
plt.figure(figsize=(3,3))

x = np.arange(5,101,5)
ours_arr = np.array(ours)
for i in range(3):
    plt.plot(x, ours_arr[:,i], '.-', linewidth=3, markersize=10) 

plt.gca().set_prop_cycle(None)
for i in range(3):
    #calculate equation for trendline
    z = np.polyfit(x, ours_arr[:,i], 1)
    p = np.poly1d(z)

    #add trendline to plot
    plt.plot(x, p(x), linewidth=2, linestyle="--")

plt.legend(['BIL-6', 'JML-4', 'ACT'])
plt.xlabel("Epoch")
plt.ylabel("KNN test set accuracy (%)")
plt.ylim(40,80)
plt.savefig('rebuttal/ours.pdf')
```
